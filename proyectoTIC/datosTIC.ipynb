```python
import pandas as pd
import glob
import os

# Obtener todos los archivos CSV del directorio data
csv_files = glob.glob('data/*.csv')

# Leer y combinar todos los archivos CSV en un solo dataframe
dataframes = []
for file in csv_files:
    df = pd.read_csv(file, sep=';')
    dataframes.append(df)

# Concatenar todos los dataframes
df_combined = pd.concat(dataframes, ignore_index=True)

print(f"Se cargaron {len(csv_files)} archivos CSV")
print(f"Dimensiones del dataframe combinado: {df_combined.shape}")
df_combined.head()
```
